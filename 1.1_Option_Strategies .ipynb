## Long Condor Strategy Using Call Options

### Definition

The Long Condor strategy is a neutral options trading strategy used when a trader expects minimal movement in the price of the underlying asset. It involves four call options with different strike prices but the same expiration date.

### Strategy Components

1. **Buy a call option** with a lower strike price \(K_1\).
2. **Sell a call option** with a higher strike price \(K_2\) but lower than \(K_3\).
3. **Sell another call option** with a higher strike price \(K_3\) but lower than \(K_4\).
4. **Buy a call option** with a higher strike price \(K_4\).

### Conditions

- \(K_1 < K_2 < K_3 < K_4\)
- All options have the same expiration date.

### Payoff Calculation

The payoff for a Long Condor spread can be calculated as follows:

\[
\text{Payoff} = 
\begin{cases} 
-(C_1 - C_2 + C_3 - C_4) & \text{if } S_T < K_1 \\
-(C_1 - C_2 + C_3 - C_4) + (S_T - K_1) & \text{if } K_1 \leq S_T < K_2 \\
C_2 - C_1 - C_4 + C_3 & \text{if } K_2 \leq S_T < K_3 \\
C_2 - C_1 - C_4 + C_3 - (S_T - K_3) & \text{if } K_3 \leq S_T < K_4 \\
-(C_1 - C_2 + C_3 - C_4) & \text{if } S_T \geq K_4
\end{cases}
\]

Where:
- \(S_T\) is the price of the underlying asset at expiration.
- \(C_1, C_2, C_3, C_4\) are the premiums paid or received for the call options.

### Profit and Loss

- **Maximum Profit:** Limited to \(C_2 - C_1 - C_4 + C_3\).
- **Maximum Loss:** Limited to the net premium paid \(-(C_1 - C_2 + C_3 - C_4)\).
- **Breakeven Points:** There are two breakeven points for this strategy:
  - Lower breakeven point: \(K_1 + (C_1 - C_2 + C_3 - C_4)\).
  - Upper breakeven point: \(K_4 - (C_1 - C_2 + C_3 - C_4)\).

### Advantages and Disadvantages

- **Advantages:**
  - Limited risk with a known maximum loss.
  - Potential to profit from a range-bound market.

- **Disadvantages:**
  - Profits are also limited.
  - Requires accurate prediction of the price range of the underlying asset.

### Conclusion

The Long Condor strategy using call options is suitable for traders who expect the price of the underlying asset to remain within a certain range. It offers a balance between risk and reward, with both potential profit and loss being limited.

## Short Condor Strategy Using Call Options

### Definition

The Short Condor strategy is a neutral options trading strategy used when a trader expects a significant movement in the price of the underlying asset, but is unsure of the direction. It involves four call options with different strike prices but the same expiration date.

### Strategy Components

1. **Sell a call option** with a lower strike price \(K_1\).
2. **Buy a call option** with a higher strike price \(K_2\) but lower than \(K_3\).
3. **Buy another call option** with a higher strike price \(K_3\) but lower than \(K_4\).
4. **Sell a call option** with a higher strike price \(K_4\).

### Conditions

- \(K_1 < K_2 < K_3 < K_4\)
- All options have the same expiration date.

### Payoff Calculation

The payoff for a Short Condor spread can be calculated as follows:

\[
\text{Payoff} = 
\begin{cases} 
(C_1 - C_2 + C_3 - C_4) & \text{if } S_T < K_1 \\
(C_1 - C_2 + C_3 - C_4) - (S_T - K_1) & \text{if } K_1 \leq S_T < K_2 \\
-(C_2 - C_1 - C_4 + C_3) & \text{if } K_2 \leq S_T < K_3 \\
-(C_2 - C_1 - C_4 + C_3) + (S_T - K_3) & \text{if } K_3 \leq S_T < K_4 \\
(C_1 - C_2 + C_3 - C_4) & \text{if } S_T \geq K_4
\end{cases}
\]

Where:
- \(S_T\) is the price of the underlying asset at expiration.
- \(C_1, C_2, C_3, C_4\) are the premiums paid or received for the call options.

### Profit and Loss

- **Maximum Profit:** Limited to the net premium received \((C_1 - C_2 + C_3 - C_4)\).
- **Maximum Loss:** Limited to \((K_3 - K_2) - (C_1 - C_2 + C_3 - C_4)\).
- **Breakeven Points:** There are two breakeven points for this strategy:
  - Lower breakeven point: \(K_1 + (C_1 - C_2 + C_3 - C_4)\).
  - Upper breakeven point: \(K_4 - (C_1 - C_2 + C_3 - C_4)\).

### Advantages and Disadvantages

- **Advantages:**
  - Limited risk with a known maximum loss.
  - Potential to profit from significant price movements in either direction.

- **Disadvantages:**
  - Profits are also limited.
  - Requires accurate prediction of the volatility of the underlying asset.

### Conclusion

The Short Condor strategy using call options is suitable for traders who expect significant volatility in the price of the underlying asset but are uncertain about the direction of the movement. It offers a balance between risk and reward, with both potential profit and loss being limited.



```python
option.short_condor_call(sigma, 80, 90, 110, 120)
```

```python
option.short_condor_call(sigma, 100, 90, 110, 120)
```

## Bull Call Spread Strategy

### Definition

The Bull Call Spread is a bullish options trading strategy used when a trader expects a moderate rise in the price of the underlying asset. It involves buying a call option with a lower strike price and selling another call option with a higher strike price, both with the same expiration date.

### Strategy Components

1. **Buy a call option** with a lower strike price \(K_1\).
2. **Sell a call option** with a higher strike price \(K_2\).

### Conditions

- \(K_1 < K_2\)
- Both options have the same expiration date.

### Payoff Calculation

The payoff for a Bull Call Spread can be calculated as follows:

\[
\text{Payoff} = 
\begin{cases} 
-C_1 + C_2 & \text{if } S_T < K_1 \\
-C_1 + C_2 + (S_T - K_1) & \text{if } K_1 \leq S_T < K_2 \\
C_2 - C_1 + (K_2 - K_1) & \text{if } S_T \geq K_2
\end{cases}
\]

Where:
- \(S_T\) is the price of the underlying asset at expiration.
- \(C_1\) is the premium paid for the call option with strike price \(K_1\).
- \(C_2\) is the premium received for the call option with strike price \(K_2\).

### Profit and Loss

- **Maximum Profit:** Limited to \(K_2 - K_1 - C_1 + C_2\).
- **Maximum Loss:** Limited to the net premium paid \(-C_1 + C_2\).
- **Breakeven Point:** The breakeven point for this strategy is \(K_1 + C_1 - C_2\).

### Advantages and Disadvantages

- **Advantages:**
  - Limited risk with a known maximum loss.
  - Potential to profit from a moderate rise in the underlying asset's price.

- **Disadvantages:**
  - Profits are limited.
  - Requires the underlying asset's price to rise to make a profit.

### Conclusion

The Bull Call Spread strategy is suitable for traders who have a bullish outlook on the underlying asset but expect only a moderate price increase. It offers a balance between risk and reward, with both potential profit and loss being limited.


```python
option.bull_call_spread(sigma, 100, 120)
```

```python
option.bull_call_spread(sigma, 100, 100)
```

## Bear Call Spread Strategy

### Definition

The Bear Call Spread is a bearish options trading strategy used when the trader expects a moderate decrease in the price of the underlying asset. It involves selling a call option with a lower strike price and buying another call option with a higher strike price, both with the same expiration date.

### Strategy Components

1. **Sell a call option** with a lower strike price \(K_1\).
2. **Buy a call option** with a higher strike price \(K_2\).

### Conditions

- \(K_1 < K_2\)
- Both options have the same expiration date.

### Payoff Calculation

The payoff for a Bear Call Spread can be calculated as follows:

\[
\text{Payoff} = 
\begin{cases} 
C_1 - C_2 & \text{if } S_T < K_1 \\
C_1 - C_2 - (S_T - K_1) & \text{if } K_1 \leq S_T < K_2 \\
-(C_2 - C_1) & \text{if } S_T \geq K_2
\end{cases}
\]

Where:
- \(S_T\) is the price of the underlying asset at expiration.
- \(C_1\) is the premium received for selling the call option with strike price \(K_1\).
- \(C_2\) is the premium paid for buying the call option with strike price \(K_2\).

### Profit and Loss

- **Maximum Profit:** Limited to the net premium received \(C_1 - C_2\).
- **Maximum Loss:** Limited to the difference between the strike prices minus the net premium received \((K_2 - K_1) - (C_1 - C_2)\).
- **Breakeven Point:** \(K_1 + (C_1 - C_2)\).

### Advantages and Disadvantages

- **Advantages:**
  - Limited risk with a known maximum loss.
  - Potential to profit from a slight decrease in the price of the underlying asset.

- **Disadvantages:**
  - Profits are also limited.
  - Not suitable for a significant bearish move in the underlying asset.

### Conclusion

The Bear Call Spread strategy is suitable for traders who have a moderately bearish outlook on the underlying asset. It offers a way to generate income through the premiums received, with a controlled level of risk.


```python
option.bear_call_spread(sigma, 100, 120)
```
